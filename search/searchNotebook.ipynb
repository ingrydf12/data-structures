 # Contexto
 
 Busca é o processo algorítmico de encontrar um **item específico** numa coleção de itens. Uma busca tipicamente devolve **True ou False**, indicando se o item está presente ou não.

 Em Python, existe um jeito fácil de verificar se um item está presente numa lista de itens. Basta usar o operador in. Mas vamos utilizar a forma sem funções nativas do Python.

# SEQUENCIAL

Quando itens de dados são armazenados numa coleção tal qual uma lista, nós dizemos que eles têm uma relação linear ou sequencial. Como os índices são ordenados, eles são sequenciais.

Começando pelo primeiro item da lista, nós simplesmente andamos de item a item, obedecendo à disposição sequencial até que encontremos o que nós estamos procurando ou que não haja mais itens a serem buscados. Quando não há mais itens, então chegamos à conclusão de que o item buscado não estava presente.

```python
# Exemplo com base no exercício de Item e Mercado

class Item:
    def __init__(self, id, descricao, preco):
        self.id = id
        self.descricao = descricao
        self.preco = preco
        self.anterior = None
        self.proximo = None
        
    def __str__(self):
        return f"Item: {self.id} | {self.descricao} / Valor: {self.preco}"

class Mercado:
    def __init__(self):
        # Indicam o ínicio (head) e fim (tail) da lista d.e
        self.head = None
        self.tail = None
        
    def adicionar_item_marketplace(self, item_id, name, price):
        novo_item = Item(item_id, name, price);
        if not self.head:
            self.head = novo_item
            self.tail = novo_item
            return
        else:
            self.tail.proximo = novo_item
            self.tail = novo_item
        # faremos a busca sequencial com base no ID do item
    def busca_sequencial(self, id):
        # variável atual pra guardar o valor de self.head pra tomar base como o início da procura
        atual = self.head
        # verificar primeiro se ele não é vazio
        if not atual:
            print("Lista está vazia")
            return
        # enquanto o atual existir e o ID dele não for igual ao que eu passei de parâmetro, ele vai printar e passar para o próximo (atualizando o valor da var atual)
        while atual and atual.id != id:
            print("Percorrendo: {atual}")
            atual = atual.proximo
        if atual.id == id:
            print(f"Está presente na lista: {atual}.")
            return


mercado = Mercado();
def main():
    print("MERCADO 1");
    print("1. Adicionar item")
    print("2. Busca sequencial");

    while True:
        try:
            escolha = int(input("Digite sua escolha: "))
        except ValueError:
            print("Entrada inválida. Por favor, digite um nÃºmero.")
            continue

        if escolha == 1:
            item_id = input("Digite o ID: ")
            nome = input("Digite o nome do item: ")
            price = input("Digite um preço para o item: ")
            mercado.adicionar_item_marketplace(item_id, nome, price)
        elif escolha == 2:
            item_id = input("Digite o ID a ser buscado: ")
            mercado.busca_sequencial(item_id);

if __name__ == "__main__":
    main()
```
